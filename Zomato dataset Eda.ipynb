```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.info()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%',radius = 1.6,)
plt.legend()
plt.show()
```

outcome: mojor bussiness of zomato is happing in india
The second highest  country to make business is United States

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

Observation: from 4.5 to 4.9 rating is excellent
from 1.8 to 2.5 rating is poor
from 3 to 3.4 rating is Average

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data= ratings,ax=ax,palette=["Blue","red","orange","yellow","green","green"])
```

Observation: count for the people who have not rated the app is much higher.
The average number of people who have rated are b/w 2.6 and 3.7
